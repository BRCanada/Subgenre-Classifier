# Preview Conversion
#### Objective
1. Iterate through new list of preview_url in tracklist_dropna.csv
2. Convert to audio files with requests
3. Save updated .csv with audio paths column
    

```python
import sys
sys.path.append('../')

from tqdm import tqdm
import re
import requests
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../data/tracklist_dropna.csv", sep=";", index_col=[0])
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data.head()
```

```python
len(data)
```

```python
data['file_location'] = ""
```

```python
data.head()
```

```python
def clip_downloader(df):
    
    
    for i in tqdm(range(len(df))):
    
        song_url = df['preview_url'][i]
        
        trackname = re.sub(r"[^a-zA-Z0-9 ]", "", df['track_name'][i])
        artistname = re.sub(r"[^a-zA-Z0-9 ]", "", df['artist'][i])
        
        local_filename = trackname + '-' + artistname
        
        headers = {'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:78.0) Gecko/20100101 Firefox/78.0'}
    
        with open(f'../data/audio/{local_filename}.wav', 'wb') as f_out:
            r = requests.get(song_url, headers=headers, stream=True)
            for chunk in r.iter_content(chunk_size=1024):
                if chunk:
                    f_out.write(chunk)
        
        data.at[i, 'file_location'] = f'../data/audio/{local_filename}.wav'
```

```python
clip_downloader(data)
```

```python
data.to_csv('../data/tracklist_mp3link.csv', sep=";", index=False)
```
